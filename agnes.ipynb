```python
import pandas as pd
import scipy.cluster.hierarchy as shc
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataset = pd.read_csv('heirarchical.csv')
labels=np.asarray(dataset.iloc[:,0])
datapoints = dataset.iloc[:,1:]
dataset
```

```python
plt.figure(figsize=(10, 7))  
plt.title("Single Linkage")   
dend = shc.dendrogram(shc.linkage(datapoints, method='single'), labels=labels)
```

```python
plt.figure(figsize=(10, 7))  
plt.title("Complete Linkage")  
dend = shc.dendrogram(shc.linkage(datapoints, method='complete'), labels=labels)
```

```python
plt.figure(figsize=(10, 7))  
plt.title("Average Linkage")  
dend = shc.dendrogram(shc.linkage(datapoints, method='average'), labels=labels)
```
