```python
import pandas as pd
```

```python
from sklearn.datasets import load_iris 
data = load_iris(as_frame=True)
df = pd.DataFrame(data.data)
```

```python
df["target"]=data.target
```

```python
df
```

```python
import numpy as np
from sklearn import tree
clf  =  tree.DecisionTreeClassifier(criterion='entropy')
img  =clf.fit(np.array(df["sepal length (cm)"]).reshape(-1,1),df["target"])
```

```python
tree.plot_tree(img,max_depth=3)
```

```python
import graphviz 
dot_data = tree.export_graphviz(img,max_depth=3) 
graph = graphviz.Source(dot_data) 
graph
```

```python
from sklearn import svm
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

data = pd.read_csv('linearsvm.csv')
X = data[['x1', 'x2']].values
y = data['y'].values
model = svm.SVC(kernel='linear')
model.fit(X, y)
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
plt.scatter(model.support_vectors_[:,0], model.support_vectors_[:,1], color='red')
```
